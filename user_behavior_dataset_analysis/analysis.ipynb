```python
group_11 = df[df['App Usage Time (min/day)'] > 500]['App Usage Time (min/day)'].reset_index(drop=True)
group_22 = df[(df['App Usage Time (min/day)'] > 250) & (df['App Usage Time (min/day)'] <= 500)]['App Usage Time (min/day)'].reset_index(drop=True)
group_33 = df[df['App Usage Time (min/day)'] <= 250] ['App Usage Time (min/day)'].reset_index(drop=True)

group_1 = len(group_11)
group_2 = len(group_22)
group_3 = len(group_33)


print(group_1)
print(group_2)
print(group_3)
```

```python
group_labels = ['+500', '250-500', '250 and less']
group_counts = [group_1, group_2, group_3]

plt.figure(figsize=(10, 6))
plt.bar(group_labels, group_counts, color=['blue', 'green', 'red'])
plt.title('App Usage Time')
plt.xlabel('min/day')
plt.ylabel('Counts')
plt.xticks(rotation=0)

```
